## Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

#### Ans.
#### ArithmeticErrors:-
#### ZeroDivisionError
#### OverflowError
#### FloatingPointError

#### Example:-

#### ---> ZeroDivisionError:- This error comes when we divide a number by 0.

#### Example:-

```python
try:
    5/0
except ArithmeticError as e:
    print('The error is:- ', e)
```

#### ---> OverflowError:- This error comes when a variable is storing value above its limit.

#### Example:-

```python
j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
except ArithmeticError as e:
    print('The error is due to overflow:- ', e)
```

## Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

#### Ans.
#### The LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively.

#### We can use LookupError exception class to handle both
#### IndexError and
#### KeyError

## IndexError Example:-

```python
try:
    l = [4,5,6,7]
    print(l[5])
except LookupError as e:
    print('The issue is:- ', e)
```

## KeyError Example:-

```python
try:
    d = {'name' : 'asd',
         'age' : 20
        }
    print(d['key'])
except LookupError as e:
    print('The issue is:- ', e, 'not exist in d.')
```
